# Generating database

```python
#make_blobs gives us fitures=x and y=lables
from sklearn.datasets import make_blobs

# As here n_features=2 we see one feature in x vector and one in y vector. 
X, y = make_blobs(n_samples=300, n_features=2)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], color= 'b')
plt.show()
# What we see as Y vector here is not the label. It's the second feature.
```

# K Means

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
model.fit(X)
```

```python
centers = model.cluster_centers_
labels = model.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c= labels)
plt.scatter(centers[:, 0], centers[: , 1], c='red', marker='x')
plt.show()
```
