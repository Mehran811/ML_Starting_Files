#Matplotlib

To draw graphs

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,8,2,5,3,6,8,1]

plt.plot(x, y, color="green")
plt.title("example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#To show cicles we use 'o'or '*'
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,8,2,5,3,6,8,1]

plt.plot(x, y, 'o', color="green")
plt.title("example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#To show both: marker='*'
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,8,2,5,3,6,8,1]

plt.plot(x, y, marker='*', color="green")
plt.title("example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Bars plt.bar
x = ["blue", "green", "yellow", "red"]
y = [2,4,8,2]

plt.bar(x, y, color="red")
plt.title("example")
plt.xlabel("color")
plt.ylabel("count")
plt.show()
```

```python
#Circle plt.pie
x = ["blue", "green", "yellow", "red"]
y = [2,4,8,2]

#To show percentage autopct="%1.1f"
#To show % as well: "%1.1f%%"
plt.pie(y, labels=x, autopct="%1.1f%%", colors=["blue", "green", "yellow", "red"])
#to define colors in order: colors=["blue", "green", "yellow", "red"]
plt.title("test")
plt.show()
```

```python
# Histogram              
# To import random numbers
import numpy as np

x = np.random.rand(1000, 1)
plt.hist(x)
plt.show()
```

```python
#Define the number of bars with bins
plt.hist(x, bins=5)
plt.show()
```

```python
# Heatmap

x2 = np.random.rand(64, 64)

plt.imshow(x2)
plt.show()
```

```python
x2 = np.random.rand(64, 64)

plt.imshow(x2, cmap="hot")
plt.show()
```

```python
x2 = np.random.rand(64, 64)

plt.imshow(x2)
plt.colorbar()
plt.show()
```
